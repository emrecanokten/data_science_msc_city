#### 5. Selecting numeric columns and mapping to [0, 1] interval

```python
df_b_f = df_b.select_dtypes(include=['float64'])
df_b_i = df_b.select_dtypes(include=['int64'])

df_n = pd.concat([df_b_i, df_b_f], axis=1)
```

```python
# drop the columns that contain numeric but not context-related
df_n = df_n.drop(['iso_numeric','year','e_prev_100k_hi_log'], axis=1)
```

```python
# replace null values with mean of corresponding column
df_n.fillna(df_n.mean(), inplace = True)
```

```python
# mapping

df_n_mapped = df_n.copy()

for cl in df_n.columns:
        mx = df_n[cl].max()
        mn = df_n[cl].min()
        df_n_mapped[cl] = (df_n[cl] - mn) / (mx - mn)
```
